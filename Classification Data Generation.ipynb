```python
"""
In this notebook, I create different versions of classification datas,
for using at other projects.I also visualize to understand the distribution.
"""
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
from sklearn.datasets import make_classification
from sklearn.datasets import make_blobs
from sklearn.datasets import make_gaussian_quantiles


```

```python
def plot_distribution(Xs,Ys,title):
    plt.figure(figsize=(8, 8))
    plt.subplots_adjust(bottom=.05, top=.9, left=.05, right=.95)

    plt.subplot(321)
    plt.title(title, fontsize='small')
    plt.scatter(Xs[:, 0], Xs[:, 1], marker='o', c=Ys,
            s=25, edgecolor='k')
    plt.show()
```

```python
X1, Y1 = make_classification(n_samples=40,n_features=2, n_redundant=0, n_informative=1,
                             n_clusters_per_class=1,flip_y=0.1)
plot_distribution(X1,Y1,"simple case")
```

```python
X, y = make_classification(n_samples=10000, n_features=25)
plot_distribution(X,y,"lots of features")
```

```python
#  values of Y will be randomly flipped
for i in np.arange(0.0,0.8,0.1):
    X, y = make_classification(n_samples=10000, n_features=25,flip_y=i)
    plot_distribution(X,y,str(i)+" flipped")
```

```python
# The lower the value, the harder classification is.
for i in np.arange(0.0,0.8,0.1):
    X, y = make_classification(n_samples=500, n_features=25,class_sep=i)
    plot_distribution(X,y,str(i)+" class seperation")
```

```python
X, y = make_classification(
        n_samples=500, n_features=25,
        n_redundant=10, # 10 out of the 25 features will just be combinations of other features 
        n_repeated=5) # and 5 out of 25 features will be duplicates)
plot_distribution(X,y,"feature contribution")
```

```python
X, y = make_classification(
        n_samples=500, n_features=25,
        n_redundant=15, # 10 out of the 25 features will just be combinations of other features 
        n_repeated=5) # and 5 out of 25 features will be duplicates)
plot_distribution(X,y,"feature contribution")
```

```python
X, y = make_classification(
        n_samples=500, n_features=25,
        n_informative=15,
        n_redundant=5, 
        n_repeated=5) 
plot_distribution(X,y,"feature contribution")
```

```python
X, y = make_classification(
        n_samples=500, n_features=25,
        n_informative=5,
        n_redundant=15, 
        n_repeated=5) 
plot_distribution(X,y,"feature contribution")
```

```python
X, y = make_classification(n_samples=10000, 
    n_features=25, 
    n_informative=10,
    n_redundant=10, 
    n_repeated=5,
    weights=[0.05,0.95], # 20% of the targets will be 0, 80% will be 1. default is 50/50    
    ) 
plot_distribution(X,y,"weights distribution")
```
